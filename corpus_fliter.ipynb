```python
import json
import os

dataset_name = "dbpedia"

input_path = "/root/autodl-tmp/LYQ/rag_evaluation/eval_data/"
output_path = "/root/autodl-tmp/LYQ/rag_evaluation/eval_data_filtered"


test_file = os.path.join(input_path,dataset_name, "qrels/test.jsonl")
other_file = os.path.join(input_path,dataset_name, "qrels/dev.jsonl")
corpus_file = os.path.join(input_path,dataset_name, "corpus.jsonl")
output_file = os.path.join(output_path,dataset_name, "corpus.jsonl")


output_dir = os.path.dirname(output_file)  
if not os.path.exists(output_dir):        
    os.makedirs(output_dir)              
    print(f"Created directory: {output_dir}")
else:
    print(f"Directory already exists: {output_dir}")

# 收集 test.jsonl 中的所有 corpus-id
def collect_corpus_ids(test_file,other_file):
    corpus_ids = set()
    with open(test_file, "r", encoding="utf-8") as f:
        for line in f:
            data = json.loads(line)
            corpus_ids.add(data["corpus-id"])

    with open(other_file, "r", encoding="utf-8") as f:
        for line in f:
            data = json.loads(line)
            corpus_ids.add(data["corpus-id"])
        
    return corpus_ids

# 过滤 corpus.jsonl
def filter_corpus(corpus_file, corpus_ids, output_file):
    with open(corpus_file, "r", encoding="utf-8") as f_in, open(output_file, "w", encoding="utf-8") as f_out:
        for line in f_in:
            data = json.loads(line)
            if data["_id"] in corpus_ids:  # 判断 corpus 的 _id 是否在 test 中出现
                f_out.write(json.dumps(data, ensure_ascii=False) + "\n")


# 收集 test 中引用的 corpus-id
corpus_ids = collect_corpus_ids(test_file,other_file)
print(f"Collected {len(corpus_ids)} corpus IDs from test.jsonl.")

# 过滤 corpus 并输出裁剪后的文件
filter_corpus(corpus_file, corpus_ids, output_file)
print(f"Filtered corpus saved to {output_file}.")


# 计算文件大小的函数
def calculate_file_size(file_path):
    return os.path.getsize(file_path)

# 示例路径
file_path =output_file # 替换为你的文件路径
file_size = calculate_file_size(file_path)


def count_lines_in_jsonl(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        line_count = sum(1 for _ in f)  # 逐行读取并计数
    return line_count


line_count = count_lines_in_jsonl(file_path)
print(f"Total number of lines: {line_count}")
print(f"File size (num_bytes): {file_size} bytes")

    
```
